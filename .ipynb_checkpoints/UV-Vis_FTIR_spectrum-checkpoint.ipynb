# Define any functions

```python
## Print the name of all files
import os

#Universal path for the working dictionary
PATH = 'data/'

files = os.listdir(PATH)
for file in files:
    print(file)
```

```python
'''numpy.loadtxt(fname, dtype=<class 'float'>, 
              comments='#',
              delimiter=None, 
              skiprows=0, 
              usecols=None, 
              unpack=False)'''


def load_d(name):
    xy_data=np.loadtxt(PATH+str(name),skiprows=3)
    x1=[]
    y1=[]
    for i in xy_data:
        x1.append(i[0])
        y1.append(float(i[1]))
    return np.array(x1), np.array(y1)
```

```python
def normal(a):
    amin, amax = min(a), max(a)
    for i, val in enumerate(a):
        a[i] = (val-amin) / (amax-amin)
    return np.array(a)
```

# Plot UV-Vis Transmittance to Absorbance

```python
def diff(x_raw,y_raw,x_bg,y_bg):
    
    x_raw_diff = x_raw-x_bg
    y_raw_diff = y_bg-y_raw
    return x_raw,y_raw_diff

def TransToAbso(x_raw,y_raw,x_bg,y_bg):
    abso_raw = []
    for i in range(len(x_raw)):
        #abso_01 = np.log10(-(y_raw[i]-y_bg[i])/100)
        abso_01 = -(np.log10(y_raw[i]/100)-np.log10(y_bg[i]/100))

        abso_raw.append(abso_01)
    return x_raw,abso_raw
```
